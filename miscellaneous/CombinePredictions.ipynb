# Combine Predictions
Quickly combine the predictions of 2 models using a weighted average.

## Import

```python
import pandas as pd
import numpy as np

#weight to use. (e.g.)
df1_weight = 10
df2_weight = 1

#df1 and df2 contain the model predictions to be combined
df1 = pd.read_csv('/Users/eloiserosen/Downloads/submission.csv')
df2 = pd.read_csv('/Users/eloiserosen/Downloads/RF.csv.gz', compression='gzip', header=0, sep=',', quotechar='"')
```

## Check that column headings and row numbers match

```python
# check column headings match
col1 = list(df1.columns.values)
col2 = list(df2.columns.values)
intersection = set(col1).intersection(col2)
intersection = list(intersection)

print len(col1)
print len(col2)
print len(intersection)

print intersection
```

```python
# check row numbers match
print len(df1.index)
print len(df2.index)
```

## Combine

```python
#new df
df3 = pd.DataFrame(data=None, columns=df1.columns,index=df1.index)
df3.head()
```

```python
#fill with weighted average
for word in col1:
    df3[word] = (df1[word]*df1_weight + df2[word]*df2_weight)/(df1_weight+df2_weight)
```

# Output CSV

```python
df3.to_csv('combined_output.csv', index=False)
```
